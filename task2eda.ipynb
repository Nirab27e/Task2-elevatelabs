```python
import pandas as pd
df=pd.read_csv(r"C:\Users\DELL\Downloads\Titanic-Dataset.csv")
print("DATA INFORMATION:")
print(df.info())
print("FIRST FEW ROWS:")
print(df.head())

```

```python
print("SUMMARY STATISTICAL DATA")
print(df.describe())
```

```python
print("Median")
print(df.median(numeric_only=True))
print("\n")

print("Mean")
print(df.mean(numeric_only=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
df[['Age','Fare','SibSp','Parch']].hist(bins=30,figsize=(12,8),edgecolor='black')
plt.suptitle("Histogram of Numeric values:",fontsize=16)
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

```python
num_col=['Age','Fare','SibSp','Parch']
for c in num_col:
    plt.figure(figsize=(9,2))
    sns.boxplot(x=df[c],color='green')
    plt.title(f'BOX PLOT OF {c}')
    plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.7)
```

```python
subset = ['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']
sns.pairplot(df[subset],hue='Survived',palette='coolwarm')
plt.suptitle("Pairplot of Features Colored by Survival", y=1.02)
plt.show()
```

```python
# Identify Patterns, Trends, Anomalies :
# Age has a wide distribution; children and elderly are present.
# Fare is highly skewed with extreme outliers.
# Most passengers had no siblings/spouses or parents/children aboard.
# Strong negative correlation between Pclass and Fare.
# Survivors are more likely in lower SibSp/Parch ranges and higher fare brackets.

# Basic Feature-Level Inferences
# Age: Median is ~28. Elderly passengers less common. Outliers above 70 exist.
# Fare: Most passengers paid < $50. Outliers paid > $300.
# SibSp/Parch: Most traveled alone. High values are rare and considered outliers.
# Pclass: 3rd class passengers were most common. Lower class = lower survival probability.
# Correlation: Fare negatively correlates with Pclass (lower class  cheaper ticket).
# Survival: More likely with higher fare, lower class, younger age.

```
